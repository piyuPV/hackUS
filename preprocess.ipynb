```python
import os
from pydub import AudioSegment
from pydub.utils import which

# Set FFmpeg path if not in system PATH
AudioSegment.converter = which("ffmpeg")

# Paths to input and output folders
input_folder = "C:\\Users\\Viraj Wadke\\Desktop\\Projects\\sandalwood\\dataset\\Dataset"
output_folder = "C:\\Users\\Viraj Wadke\\Desktop\\Projects\\sandalwood\\wavd"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Loop through all MP3 files in the input folder
for file_name in os.listdir(input_folder):
    if file_name.endswith(".mp3"):
        # Construct full file paths
        mp3_path = os.path.join(input_folder, file_name)
        wav_path = os.path.join(output_folder, os.path.splitext(file_name)[0] + ".wav")
        
        try:
            # Convert MP3 to WAV
            audio = AudioSegment.from_mp3(mp3_path)
            audio.export(wav_path, format="wav")
            print(f"Converted: {file_name} to {os.path.basename(wav_path)}")
        except Exception as e:
            print(f"Failed to convert {file_name}: {e}")

print("All files have been converted!")
```

```python
import os
import librosa
import noisereduce as nr
import soundfile as sf

# Paths to input and output folders
input_folder = "C:\\Users\\Viraj Wadke\\Desktop\\Projects\\sandalwood\\wavd"
output_folder = "C:\\Users\\Viraj Wadke\\Desktop\\Projects\\sandalwood\\cleaned_wavd"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Preprocessing each audio file
for file_name in os.listdir(input_folder):
    if file_name.endswith(".wav"):
        input_path = os.path.join(input_folder, file_name)
        output_path = os.path.join(output_folder, file_name)
        
        try:
            # Load the audio file
            audio, sr = librosa.load(input_path, sr=None)
            
            # Estimate noise profile from the initial part of the audio
            noise_sample = audio[:sr]  # Use the first second for noise profile
            
            # Apply noise reduction
            reduced_noise = nr.reduce_noise(y=audio, sr=sr, y_noise=noise_sample)
            
            # Save the cleaned audio file
            sf.write(output_path, reduced_noise, sr)
            
            print(f"Processed and saved: {file_name}")
        except Exception as e:
            print(f"Failed to process {file_name}: {e}")

print("Preprocessing complete!")
```
